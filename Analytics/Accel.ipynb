```python
# filename = "rawaccel_move_test.csv"
filename = "rawaccel_move_test_topofMBA.csv"
```

```python
with open(filename) as f:
    datur = f.readlines()
for i,line in enumerate(datur):
    datur[i] = line.replace("\n", "")
    
datas = []
counter = -1
for i in range(len(datur)-1):
    if datur[i] == "" and datur[i+1] != "":
        counter += 1
    elif datur[i] == "" and datur[i+1] == "":
        continue
    else:
        val = [float(d) for d in datur[i].split(",")] + [counter]
        datas.append(val)
        
full_df = pd.DataFrame(datas, columns=["x","y","z","t","run"])

# Adjust the timestamps
for i in full_df.run.unique():
    full_df[full_df.run == i].t -= full_df[full_df.run == i].t.values[0]
```
